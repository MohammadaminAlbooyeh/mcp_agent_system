# Multi-Step Reasoning

This notebook explores the agent's reasoning capabilities including chain-of-thought, ReAct patterns, and reflection.

## Reasoning Modules

1. **ChainOfThought** - Step-by-step reasoning
2. **ReActPattern** - Reasoning + Acting loop
3. **Reflection** - Self-evaluation and correction
4. **ErrorRecovery** - Handling and recovering from failures
5. **DecisionMaker** - Ranking and selecting options

```python
# Explore reasoning modules
from agent.reasoning.chain_of_thought import ChainOfThought
from agent.reasoning.react_pattern import ReActPattern
from agent.reasoning.reflection import Reflection
from agent.reasoning.decision_maker import DecisionMaker
from agent.core.agent import Agent

agent = Agent({"llm": "openai"})

cot = ChainOfThought(agent)
print(f"ChainOfThought: {type(cot).__name__}")

react = ReActPattern(agent)
print(f"ReActPattern: {type(react).__name__}")

reflection = Reflection(agent)
print(f"Reflection: {type(reflection).__name__}")

dm = DecisionMaker(agent)
print(f"DecisionMaker: {type(dm).__name__}")
```

```python
# Decision making demo
import asyncio

async def test_decision():
    agent = Agent({"llm": "openai"})
    dm = DecisionMaker(agent)
    options = [
        {"name": "Web Search", "description": "Search the web for information"},
        {"name": "Database Query", "description": "Query local database"},
        {"name": "File Read", "description": "Read from file system"},
    ]
    result = await dm.rank_options(options, "speed")
    print(f"Ranked options: {result}")

asyncio.run(test_decision())
```

The reasoning modules work together to break down complex tasks into manageable steps.
