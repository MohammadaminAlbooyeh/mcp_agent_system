# Agent with MCP

This notebook shows how the agent connects to the MCP server to use tools and resources.

## Agent-MCP Integration

The agent uses `MCPClient` to connect to the MCP server. Through this connection, the agent can:
- List available tools
- Call tools with arguments
- Read resources
- Get structured results

```python
# Agent MCP client setup
from agent.mcp_client.client import MCPClient

client = MCPClient()
print(f"MCP Client created (not connected)")
print(f"  Server URL: {client.server_url}")
print(f"  Session: {client.session}")
```

```python
# Connection manager for multi-server setups
from agent.mcp_client.connection_manager import ConnectionManager

cm = ConnectionManager()
print(f"Connection manager created")
print(f"  Connections: {cm.connections}")
print(f"  Active: {cm.active_connection}")
```

```python
# Tool caller wrapper
from agent.mcp_client.tool_caller import ToolCaller

caller = ToolCaller(client)
print(f"Tool caller created for client")
print(f"  Ready to call tools: calculator, web_search, etc.")
```

```python
# Complete agent with MCP integration
from agent.core.agent import Agent

agent = Agent({"llm": "openai"})
print(f"Agent with MCP client ready")
print(f"  MCP Client: {type(agent.mcp_client).__name__}")
print(f"  Can list tools: {hasattr(agent, 'get_available_tools')}")
```

The `get_available_tools()` method delegates to `mcp_client.list_tools()` to discover available capabilities.
