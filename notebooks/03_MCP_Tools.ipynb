# MCP Tools

This notebook explores all available MCP tools, their input schemas, and how to use them.

## Tool Categories

- **Web Tools**: Search, scrape, fetch URLs
- **Database Tools**: Execute SQL, read/write data
- **File Tools**: Read, write, manage files
- **Email Tools**: Read, send, parse emails
- **API Tools**: HTTP requests, REST calls, webhooks
- **Code Tools**: Execute, analyze, test code
- **Utility Tools**: Calculator, datetime, text processing

```python
# List all tools with their schemas
from mcp_server.server import MCPServer

server = MCPServer()
for name, tool in server.tools.items():
    print(f"\n=== {name} ===")
    print(f"  Description: {tool.description}")
    required = tool.input_schema.get("required", [])
    props = tool.input_schema.get("properties", {})
    for pname, pdetails in props.items():
        req = "(required)" if pname in required else "(optional)"
        print(f"  - {pname}: {pdetails.get('type', 'any')} {req}")
```

```python
# Test utility tools
from mcp_server.tools.utility_tools.calculator import CalculatorTool
from mcp_server.tools.utility_tools.datetime_tool import DateTimeTool
from mcp_server.tools.utility_tools.text_processor import TextProcessorTool

calc = CalculatorTool()
dt = DateTimeTool()
txt = TextProcessorTool()

import asyncio
async def test_tools():
    print(await calc.execute(expression="42 * 2"))
    print(await dt.execute(operation="now"))
    print(await txt.execute(text="hello mcp", operation="uppercase"))

asyncio.run(test_tools())
```

Each tool follows the same pattern: subclass `BaseTool`, define `name`, `description`, and `input_schema`, then implement `async def execute(**kwargs)`.
