# MCP Server Setup

This notebook covers how to configure and run the MCP server with different transports and settings.

## Server Configuration

The MCP server supports two transport modes:
- **stdio**: Standard input/output (default, for local/cli use)
- **http**: HTTP server (for remote access)

````python
# View current server configuration
from mcp_server.utils.config import MCPServerConfig

config = MCPServerConfig()
print(f"Transport: {config.transport}")
print(f"Host: {config.host}")
print(f"Port: {config.port}")
print(f"Log Level: {config.log_level}")
````

````python
# Start the MCP server in stdio mode (in-process demo)
# This shows how the server initializes without actually running the event loop
from mcp_server.server import MCPServer

server = MCPServer()
print(f"MCP Server '{server.server.name}' initialized")
print(f"Ready with {len(server.tools)} tools across {len(server.resources)} resources")
````

````python
# Test tool discovery
tools = [t.to_mcp_tool() for t in server.tools.values()]
for tool in tools[:5]:
    print(f"  - {tool.name}: {tool.description}")
print(f"... and {len(tools) - 5} more")
````

## Running the Server

To run the server from the command line:
```bash
# stdio mode (default)
python -m mcp_server.server

# HTTP mode
TRANSPORT=http python -m mcp_server.server
```
