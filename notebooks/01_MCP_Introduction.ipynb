# MCP Introduction

Welcome to the MCP Agent System. This notebook introduces the Model Context Protocol (MCP) and how it enables AI agents to interact with external tools and resources.

## What is MCP?

The Model Context Protocol (MCP) is an open standard that defines how AI applications (hosts) can discover and call tools, read resources, and interact with prompts provided by servers.

## Key Concepts

- **Tools**: Actions an agent can invoke (web search, database queries, file operations)
- **Resources**: Data sources the agent can read (database schemas, file contents, API responses)
- **Prompts**: Pre-defined templates for common interactions
- **Transport**: How the client and server communicate (stdio or HTTP)
