# Real-World Examples

This notebook demonstrates practical use cases for the MCP Agent System.

## Example: Research Assistant

An agent that searches the web, summarizes findings, and stores results.

```python
# Research workflow example
from agent.workflows.research_workflow import ResearchWorkflow
from agent.core.agent import Agent

agent = Agent({"llm": "openai"})
workflow = ResearchWorkflow(agent)
print(f"Research workflow initialized")
print(f"  Agent: {type(workflow.agent).__name__}")
print(f"  Ready to research topics")
```

## Example: Data Analysis

An agent that queries databases, processes data, and generates insights.

```python
# Analysis workflow example
from agent.workflows.analysis_workflow import AnalysisWorkflow

workflow = AnalysisWorkflow(agent)
print(f"Analysis workflow initialized")
print(f"  Ready to analyze data")
```

## Example: Task Automation

An agent that automates repetitive tasks like file processing and email handling.

```python
# Task automation workflow example
from agent.workflows.task_automation_workflow import TaskAutomationWorkflow

workflow = TaskAutomationWorkflow(agent)
print(f"Task automation workflow initialized")
print(f"  Ready to automate tasks")
```

## Using the Orchestrator

The `Orchestrator` manages all workflows and routes tasks to the appropriate workflow.

```python
# Orchestrator demo
from agent.core.orchestrator import Orchestrator

orch = Orchestrator({"llm": "openai"})
orch.register_workflow("research", ResearchWorkflow(orch.agent))
orch.register_workflow("analysis", AnalysisWorkflow(orch.agent))
orch.register_workflow("automation", TaskAutomationWorkflow(orch.agent))
print(f"Orchestrator with {len(orch.workflows)} workflows: {list(orch.workflows.keys())}")
```
