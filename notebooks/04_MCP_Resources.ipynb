# MCP Resources

Resources provide read-only access to data sources. The MCP Agent System includes four resource types.

## Available Resources

1. **Database Resource** (`database://schema`) - Database schema information
2. **File Resource** (`file://`) - File system access
3. **API Resource** (`api://`) - External API data
4. **Memory Resource** (`memory://state`) - Agent memory state

```python
# Explore available resources
from mcp_server.server import MCPServer

server = MCPServer()
for uri, resource in server.resources.items():
    print(f"\n=== {uri} ===")
    print(f"  Name: {resource.name}")
    print(f"  Description: {resource.description}")
    print(f"  MIME Type: {resource.mime_type}")
```

```python
# Read resources (in-process demo)
import asyncio

async def read_resources():
    for uri, resource in server.resources.items():
        print(f"\n--- {uri} ---")
        try:
            content = await resource.read()
            print(content[:500] + ("..." if len(content) > 500 else ""))
        except Exception as e:
            print(f"Error: {e}")

asyncio.run(read_resources())
```

Resources implement a simple `read()` method and can be extended by subclassing `BaseResource`.
