# Memory Management

The agent includes a sophisticated memory system with short-term, long-term, conversation history, and context window management.

## Memory Types

1. **ShortTermMemory** - Working memory for current task
2. **LongTermMemory** - Persistent storage (JSON-backed)
3. **ConversationHistory** - Sequential conversation tracking
4. **ContextWindow** - Token-aware context management

```python
# Explore memory modules
from agent.memory.short_term_memory import ShortTermMemory
from agent.memory.long_term_memory import LongTermMemory
from agent.memory.conversation_history import ConversationHistory
from agent.memory.context_window import ContextWindow

stm = ShortTermMemory()
ltm = LongTermMemory()
ch = ConversationHistory(max_length=10)
cw = ContextWindow(max_tokens=8000)

print(f"ShortTermMemory: ready")
print(f"LongTermMemory: storage={ltm.storage_path}")
print(f"ConversationHistory: max_length={ch.max_length}")
print(f"ContextWindow: max_tokens={cw.max_tokens}")
```

```python
# Memory operations demo
stm.store("key1", "value1")
stm.store("key2", "value2")
print(f"Retrieve key1: {stm.retrieve('key1')}")
print(f"All items: {stm.get_all()}")
stm.clear()
print(f"After clear: {stm.get_all()}")
```

```python
# Conversation history demo
ch.add({"role": "user", "content": "Hello"})
ch.add({"role": "assistant", "content": "Hi there!"})
ch.add({"role": "user", "content": "What can you do?"})
ch.add({"role": "assistant", "content": "I can search the web, query databases, and more!"})
print(f"Conversation length: {len(ch)}")
for msg in ch.get_history():
    print(f"  [{msg['role']}]: {msg['content']}")
```

```python
# Context window tracking
cw.add("short text", token_count=5)
cw.add("longer text example here", token_count=10)
usage = cw.get_usage()
print(f"Context usage: {usage['used']}/{usage['max']} tokens ({usage['remaining']} remaining)")
```

The `MemoryManager` class coordinates all memory modules for the agent.
