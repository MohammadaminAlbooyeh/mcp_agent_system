# Agent Basics

This notebook introduces the core Agent class, its architecture, and how it processes tasks.

## Agent Architecture

The agent follows a plan-execute-evaluate cycle:
1. **Planner** - Creates a step-by-step plan
2. **Executor** - Executes each step
3. **Evaluator** - Evaluates the results

With supporting modules:
- **Memory Manager** - Short-term and long-term memory
- **Reasoning** - Chain-of-thought, ReAct pattern
- **LLM** - Multi-provider support (OpenAI, Claude, Groq, local)

```python
# Initialize and inspect the agent
from agent.core.agent import Agent

agent = Agent({"llm": "openai"})
print(f"Agent initialized")
print(f"  Planner: {type(agent.planner).__name__}")
print(f"  Executor: {type(agent.executor).__name__}")
print(f"  Evaluator: {type(agent.evaluator).__name__}")
print(f"  Memory: {type(agent.memory).__name__}")
print(f"  Reasoning: {type(agent.reasoning).__name__}")
print(f"  LLM: {type(agent.llm).__name__}")
```

```python
# Agent memory demo
agent.memory.short_term.store("user_query", "What is AI?")
agent.memory.long_term.store("previous_topic", "Machine Learning")
print(f"Short-term: {agent.memory.short_term.retrieve('user_query')}")
print(f"Long-term: {agent.memory.long_term.retrieve('previous_topic')}")
print(f"All short-term: {agent.memory.short_term.get_all()}")
```

## Running a Task

To run a task: `await agent.run("your task here")`

The agent will plan, execute, and evaluate automatically.
